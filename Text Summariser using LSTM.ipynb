```python
import requests
import json
import pandas as pd

# Importing dataset
url = "https://datasets-server.huggingface.co/first-rows?dataset=cnn_dailymail&config=1.0.0&split=train"
response = requests.get(url)
data = json.loads(response.text)

data
```

```python
# Create empty lists for articles and highlights
articles = []
highlights = []

# Process the dataset
for item in data:
    if 'article' in item and 'highlights' in item:
        article = item['article']
        highlight = item['highlights']
        articles.append(article)
        highlights.append(highlight)

# Create a pandas DataFrame
df = pd.DataFrame({'article': articles, 'highlight': highlights})

# Print the DataFrame
print(df.head())
```
